```python
import numpy as np

# Apply log transformation to the selected columns
# You can apply log1p (log(1+x)) to handle zeros and negative values
columns_to_transform = ['qty', 'total_price', 'freight_price', 'unit_price', 
                        'product_weight_g', 'customers', 'volume', 
                        'comp_1', 'ps1', 'fp1', 'comp_2', 'ps2', 'fp2', 
                        'comp_3', 'ps3', 'fp3', 'lag_price']

# Apply log1p (log(1 + x)) to avoid issues with zero values
dataset[columns_to_transform] = dataset[columns_to_transform].apply(lambda x: np.log1p(x))

# Check the results
dataset[columns_to_transform].describe()
```

The output from the describe() function after applying the log transformation provides several key statistics for each of the transformed features. Let's go through the key points:

1. Interpretation of Transformed Values:

Mean: The mean values of the features are now lower and closer to 1, which is expected after applying a log transformation. The log transformation compresses the data range, pulling large values closer to 1 and reducing the impact of outliers.

Standard Deviation (std): The standard deviation has decreased significantly for most features, indicating that the data spread has been compressed. This is a typical outcome of log transformations, as it reduces the variance and brings the data closer to a normal distribution.

Minimum and Maximum Values: The range between the minimum and maximum values has also decreased, reflecting the reduced effect of extreme values (outliers).

2. Impact of Transformation:

Handling Zero Values: For features like freight_price, where the minimum value is zero (indicated as 0.000000 before transformation), the log transformation using log1p (which adds 1 before taking the log) handled this well by shifting zero values to a small positive number.

Normalizing Skewed Data: If the original data was heavily skewed (which is common in pricing, quantities, and weights), the log transformation would have normalized it, making the distribution more symmetrical and thus better suited for many statistical methods and machine learning models.

A. Revisualize the Data:

Plot histograms or boxplots for the transformed features to visually confirm the normalization effect.
Compare these visualizations with the original data plots to ensure that the transformation was effective.

Formula: Margin (%) = (Profit / Revenue) × 100

Purpose: 
The margin percentage indicates the profitability relative to the revenue. It shows what proportion of the revenue is retained as profit.

Why Calculate Margin?

Performance Indicator:

Margin provides insight into how efficiently a business is generating profit from its revenue. A higher margin means more profit per unit of revenue, indicating better financial health.


Comparison and Benchmarking:

Margin allows for comparison against industry benchmarks or competitors. It helps evaluate how well a company is performing relative to others in the same industry.

```python
# Calculate revenue
dataset['revenue'] = dataset['qty'] * dataset['total_price']

# Calculate profit
dataset['profit'] = dataset['revenue'] - dataset['freight_price']

# Calculate margin
dataset['margin'] = (dataset['profit'] / dataset['revenue']) * 100

```
